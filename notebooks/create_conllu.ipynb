```python
from polars import col
from stanza import Document, Pipeline
from stanza.pipeline.core import CoNLL
from label_legends.preprocess import load_data
from label_legends.util import CONLL_DIR
```

```python
df = load_data()

series = list(df.select(col("text")).collect().to_series())
series
```

```python
nlp = Pipeline('en', processors='tokenize,lemma,pos')
```

```python
documents_in = [Document([], text=d) for d in series]
# "\n\n".join(series[:100])

docs = nlp(documents_in)
# docs[0]
```

```python
CONLL_DIR.mkdir(parents=True, exist_ok=True)
for i, doc in enumerate(docs):
    CoNLL.write_doc2conll(doc, str(CONLL_DIR / f"{i}.conll"))
# preprocess(df)
```
